```python
import pandas as pd

# 1. Force ingestion of the newly exported individual customer view
df = pd.read_csv(r"C:\Users\ADMIN\Desktop\Sentinel_Internship Project1_Social Prachar\customer_risk_scores.csv")

# 2. Print structural characteristics to guarantee it's locked in
print("=== CELL 1 INGESTION VERIFICATION ===")
print("Loaded Columns:", df.columns.tolist())
print("Total Records Loaded:", len(df))
```

```python
# 1. Define Regulatory Risk Tiers based on our weighted scoring model
def assign_risk_tier(score):
    if score >= 70:
        return "Immediate Escalation (SAR Candidate)"
    elif score >= 50:
        return "Enhanced Due Diligence (EDD) Required"
    elif score >= 20:
        return "Medium Risk (Standard Monitoring)"
    else:
        return "Low Risk (Automated Clearance)"

# 2. Execute vector transformation
df['regulatory_risk_tier'] = df['customer_risk_score'].apply(assign_risk_tier)

# 3. Print the top of our newly engineered table view
print("=== CELL 2: TIERS SEGMENTATION MAPPED ===")
print(df[['customer_id', 'customer_risk_score', 'regulatory_risk_tier']].head(10))
```

```python
import numpy as np

# 1. Calculate foundational statistical metrics across all 1,000 accounts
mean_score = df['customer_risk_score'].mean()
std_score = df['customer_risk_score'].std()
upper_bound = mean_score + (2 * std_score)

print("=== FORTIS COMPLIANCE SYSTEM: DAY 10 STATISTICAL ANALYSIS ===")
print(f"System Baseline Mean Risk Score: {mean_score:.2f}")
print(f"Standard Deviation: {std_score:.2f}")
print(f"Statistical Anomaly Threshold (>2 Std Dev): {upper_bound:.2f}")

# 2. Get the breakdown counts for every single regulatory tier
print("\n=== SYSTEM RISK TIER SUMMARY ===")
print(df['regulatory_risk_tier'].value_counts())

# 3. Export this final engineered view for your Day 11 modeling
df.to_csv(r"C:\Users\ADMIN\Desktop\Sentinel_Internship Project1_Social Prachar\day10_compliance_features.csv", index=False)
print("\n[SUCCESS] Day 10 features exported to 'day10_compliance_features.csv'")
```

```python
# DAY 11: TIME-SERIES SEASONALITY & TEMPORAL ANOMALIES
# =======================================================
import numpy as np
import pandas as pd
```

```python
# 1. Generate synthetic temporal markers to isolate seasonality across the 1,000 records
np.random.seed(42)
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['transaction_day'] = np.random.choice(days_of_week, size=len(df), p=[0.15, 0.16, 0.14, 0.15, 0.25, 0.08, 0.07])
```

```python
# 2. Identify transaction frequency spikes per weekday
print("=== DAY 11: TRANSACTION FREQUENCY BY WEEKDAY ===")
weekday_counts = df['transaction_day'].value_counts().reindex(days_of_week)
print(weekday_counts)
```

```python
# 3. Detect Global Temporal Anomalies (High Volume on low-traffic weekend velocity)
print("\n=== DETECTING GLOBAL TEMPORAL ANOMALIES ===")
weekend_anomalies = df[(df['transaction_day'].isin(['Saturday', 'Sunday'])) & (df['customer_risk_score'] >= 70)]
print(f"Total Weekend High-Risk Anomalies Detected: {len(weekend_anomalies)}")
print(weekend_anomalies[['customer_id', 'transaction_day', 'total_volume', 'customer_risk_score']].head())
```

```python
# 4. Save temporal features
df.to_csv(r"C:\Users\ADMIN\Desktop\Sentinel_Internship Project1_Social Prachar\day11_time_series_features.csv", index=False)
print("\n[SUCCESS] Day 11 temporal features exported successfully!")
```
